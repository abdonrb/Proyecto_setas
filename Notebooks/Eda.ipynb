```python
distribucion_por_clase = df[df["stalk-root"] != 0].groupby("clase")["stalk-root"].value_counts(normalize=True)

def imputar_stalk_root(row):
    if row["stalk-root"] == 0:
        clase = row["clase"]
        opciones = distribucion_por_clase[clase].index
        probabilidades = distribucion_por_clase[clase].values
        return np.random.choice(opciones, p=probabilidades)
    else:
        return row["stalk-root"]

df["stalk-root"] = df.apply(imputar_stalk_root, axis=1)
# En esta funcion miramos la distribucion dentro de cada clase e imputamos aleatoriamente.
```

```python
df['stalk-root'].value_counts()
```

```python
from scipy.stats import chi2_contingency

chi2_results = {}

for col in df.columns:
    if col != "clase":
        tabla = pd.crosstab(df[col], df["clase"])
        chi2_stat, p_val, dof, expected = chi2_contingency(tabla)
        
        chi2_results[col] = {"chi2": chi2_stat, "p-valor": p_val}

chi2_df = pd.DataFrame.from_dict(chi2_results, orient="index")
chi2_df.sort_values(by="p-valor", inplace=True)
print(chi2_df)
```

#### Vemos que la variable 'odor' tiene el menor p-valor, por lo tanto es la más significativa para predecir si una seta es venenosa o no, por lo contrario veil-type es la que menos relacion tiene con "clase". Como solo tiene un unico valor vamos a eliminarla.
